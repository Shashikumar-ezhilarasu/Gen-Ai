```python
# Colab install notes (uncomment if running in Colab)
# !pip install -q diffusers transformers accelerate ftfy safetensors

from diffusers import StableDiffusionPipeline
import torch
from PIL import Image


def run_demo():
    model_id = "runwayml/stable-diffusion-v1-5"
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    pipe = pipe.to("cuda")
    prompt = "A minimal, modern illustration of a robot teaching a class, flat vector style"
    image = pipe(prompt, num_inference_steps=20, guidance_scale=7.5).images[0]
    out_path = "../examples/outputs/demo_image_example.png"
    image.save(out_path)
    print(f"Saved example to {out_path}")

if __name__ == '__main__':
    run_demo()
```
